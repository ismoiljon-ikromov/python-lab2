# 2. use BeautifulSoup to parse the page

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#prettify is method to print object's content in more readable way
print(soup.prettify())
```

```python
#children is giving you all the nested elements from the html tree
list(soup.children)
```

```python
[type(item) for item in list(soup.children)]
```

```python
#we are taking only html elements which are bringing some valuable content
#we provide over here the index for Tag element above
html = list(soup.children)[1]
list(html.children)
```

```python
body = list(html.children)[0]
```

```python
list(body.children)
```

```python
soup.find_all('a')
```

```python
#let's filter those links... those "a" elements
# this needs to be some unique value/modifier (class or id or attribute)
# that can help us distinguish the content we are looking for
cnt = soup.find_all('a', class_='posting-list-item')

cnt
```

```python
for e in cnt:
    print(e, end="\n-------------------\n"*2)
```

```python
for e in cnt:
    title = e.find("span", class_="posting-title__company")
    print(title.text.strip())
    salary = e.find("span", class_="salary")
    print(salary.text.strip())
    print("-----")
```

# 3. put data into pandas

```python
titles=[]
salaries=[]

for e in cnt:
    titles.append([title.text.strip() for title in e.find("span", class_="posting-title__company")])
    salaries.append([salary.text.strip() for salary in e.find("span", class_="salary")])
```

```python
titles
```

```python
salaries
```

```python
import pandas as pd
```

```python
jobs = pd.DataFrame({
    "title": titles,
    "salaries": salaries
})
```
